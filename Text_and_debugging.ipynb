```python
#W12a8
def check(s):
    if "." not in s:
        return False

    i = s.rfind(".")
    k = len(s) - i - 1

    if k != 3:
        return False

    ans = s[i+1:]
    if ans == "txt" or ans == "zip" or ans == "TXT" or ans == "ZIP":
        return True

    return False
class MyError(Exception):
  pass
s = input()
try:
  if check(s):
      print("Doc file thanh cong")
  else :
      raise MyError("File ko hop le")
except MyError as e :
  print(e)

```
