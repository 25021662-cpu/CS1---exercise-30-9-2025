```python
#W8a1
pi = 3.14
class Cylinder:
  def __init__ (self,radius,height):
    self.radius = radius
    self.height = height
  def total_area (self):
    return 2 * pi * self.radius * self.height + 2*pi * self.radius * self.radius
  def vol(self):
    return self.height * pi * self.radius * self.radius
r = int(input())
h = int(input())
t = Cylinder(r,h)
print(f"{t.total_area():.2f}")
print(f"{t.vol():.2f}")
```

```python
#w8a2
class Date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year

    def is_leap_year(self):
        return self.year % 400 == 0 or (self.year % 4 == 0 and self.year % 100 != 0)

    def is_valid(self):
        if self.month < 1 or self.month > 12:
            return False

        days_in_month = [31, 28, 31, 30, 31, 30,
                         31, 31, 30, 31, 30, 31]

        if self.is_leap_year():
            days_in_month[1] = 29

        if self.day < 1 or self.day > days_in_month[self.month - 1]:
            return False

        return True

    def next_day(self):
        days_in_month = [31, 28, 31, 30, 31, 30,
                         31, 31, 30, 31, 30, 31]

        if self.is_leap_year():
            days_in_month[1] = 29

        self.day += 1
        if self.day > days_in_month[self.month - 1]:
            self.day = 1
            self.month += 1
            if self.month > 12:
                self.month = 1
                self.year += 1

s = input().strip()
d, m, y = map(int, s.split('/'))

date = Date(d, m, y)

if not date.is_valid():
    print("INVALID")
else:
    print(date.next_day())

```

```python
#W8a3
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def is_O(self):
        if self.x == 0 and self.y == 0:
            return "coordinate origin"
        return None

    def axis(self):
        if self.x == 0 and self.y != 0:
            return "coordinate y axis"
        elif self.y == 0 and self.x != 0:
            return "coordinate x axis"
        return None

    def leight(self):
        import math
        return math.sqrt(self.x**2 + self.y**2)

x, y = map(int, input().split())
p = Point(x, y)

if p.is_O() is not None:
    print(p.is_O())
elif p.axis() is not None:
    print(p.axis())
    print(f"{p.leight():.2f}")
else:
    print(f"{p.leight():.2f}")
```

```python
#W8a4
class SetNumber:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2

    def set_numbers(self, num1, num2):
        self.num1 = num1
        self.num2 = num2


class Calculator(SetNumber):
    def __init__(self, num1, num2):
        super().__init__(num1, num2)

    def add(self):
        return self.num1 + self.num2

    def subtract(self):
        return self.num1 - self.num2

    def product(self):
        return self.num1 * self.num2

    def divide(self):
        if self.num2 == 0:
            return "Cannot divide by zero"
        return round(self.num1 / self.num2, 2)

    def power(self):
        return self.num1 ** self.num2

    def mod(self):
        if self.num2 == 0:
            return "Cannot divide by zero"
        return self.num1 % self.num2


a, b = map(int, input("Nhập 2 số: ").split())
cal = Calculator(a, b)

while True:
    print("Add:", cal.add())
    print("Subtract:", cal.subtract())
    print("Product:", cal.product())
    print("Divide:", cal.divide())
    print("Power:", cal.power())
    print("Mod:", cal.mod())

    exit_choice = input("Bạn có muốn thoát? ")
    if exit_choice.lower() == "yes":
        print("Thoát chương trình!")
        break
```

```python
#w8a7
import math
from math import gcd
class Fraction:
  def __init__(self,t1,m1):
    self.t1=t1
    self.m1=m1
  def KC(self,t2,m2):
    self.t2 = t2
    self.m2 = m2
  def add(self):
    if (self.m2 == self.m1):
      su = self.t1 + self.t2
      if (gcd(su,self.m1) > 1 ):
         return f"{su/gcd(su,self.m1)}/{self.gcd(su,self.m1)}"
    else :
      k = gcd(self.m1,self.m2)
      h = self.m1 // k
      g = self.m2 // k
      su = h * self.t2 + g * self.t1
      if (gcd(su,h) > 1 ):
         return f"{su/gcd(su,h)}/{h/gcd(su,h)}"
  def sub(self):
    if (self.m2 == self.m1):
      su = self.t1 - self.t2
      if (gcd(su,self.m1) > 1 ):
         return f"{su/gcd(su,self.m1)}/{self.gcd(su,self.m1)}"
    else :
      k = gcd(self.m1,self.m2)
      h = self.m1 // k
      g = self.m2 // k
      su = h * self.t2 - g * self.t1
      if (gcd(su,h) > 1 ):
         return f"{su/gcd(su,h)}/{h/gcd(su,h)}"
  def mul(self):
    su = self.t1 * self.t2
    h = self.m1 * self.m2
    if (gcd(su,h)>1) :
      return f"{su/gcd(su,h)}/{h/gcd(su,h)}"
  def div(self):
    su = self.t1 * self.m2
    h = self.t2 * self.m1
    if (gcd(su,h)>1) :
      return f"{su/gcd(su,h)}/{h/gcd(su,h)}"
a,b = map(int,input().split())
op = input()
c,d = map(int,input().split())
f1 = Fraction(a,b)
f1.KC(c,d)
if (op == "+"):
  print(f1.add())
if (op == "-"):
  print(f1.sub())
if (op == "*"):
  print(f1.mul())
if (op == "/"):
  print(f1.div())



```

```python
# W8a9
class Student:
    def __init__(self, name):
        self.name = name

class Name(Student):
    def __init__(self, name, num):
        super().__init__(name)
        self.num = num
        self.subject = {}
        self.sum = 0.0

    def add_score(self, subject, score):
        self.subject[subject] = score
        self.sum += score

    def avg(self):
        self.av = self.sum / self.num
        return f"{self.av:.2f}"

    def rank(self):
        if self.av >= 8:
            return "Excellent"
        elif self.av >= 6.5:
            return "Good"
        elif self.av >= 5:
            return "Average"
        else:
            return "Poor"

name = input()
num = int(input())

s = Name(name, num)

for _ in range(num):
    subject, score = input().split()
    s.add_score(subject, float(score))
print(name,s.avg(),s.rank())










```

```python
#w8a8
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = float(balance)

    def deposit(self, x):
        self.balance += x

    def withdraw(self, x):
        if x <= self.balance:
            self.balance -= x

    def __str__(self):
        return f"{self.owner}: {self.balance:.2f}"
owner, balance = input().split()
acc = BankAccount(owner, float(balance))

n = int(input())

for _ in range(n):
    cmd, x = input().split()
    x = float(x)

    if cmd == "DEPOSIT":
        acc.deposit(x)
    elif cmd == "WITHDRAW":
        acc.withdraw(x)

print(f"{acc.balance:.2f}")

```

```python
#w8a11
import math

class PHI:
    def __init__(self,a1,a2,a3,b1,b2,b3,c1,c2,c3,d1,d2,d3):
        self.a1=a1
        self.a2=a2
        self.a3=a3
        self.b1=b1
        self.b2=b2
        self.b3=b3
        self.c1=c1
        self.c2=c2
        self.c3=c3
        self.d1=d1
        self.d2=d2
        self.d3=d3

    def n1(self):
        self.ab1 = self.b1 - self.a1
        self.ab2 = self.b2 - self.a2
        self.ab3 = self.b3 - self.a3

        self.ac1 = self.c1 - self.a1
        self.ac2 = self.c2 - self.a2
        self.ac3 = self.c3 - self.a3

        self.n1x = self.ab2*self.ac3 - self.ab3*self.ac2
        self.n1y = self.ab3*self.ac1 - self.ab1*self.ac3
        self.n1z = self.ab1*self.ac2 - self.ab2*self.ac1

    def n2(self):
        self.bc1 = self.c1 - self.b1
        self.bc2 = self.c2 - self.b2
        self.bc3 = self.c3 - self.b3

        self.bd1 = self.d1 - self.b1
        self.bd2 = self.d2 - self.b2
        self.bd3 = self.d3 - self.b3

        self.n2x = self.bc2*self.bd3 - self.bc3*self.bd2
        self.n2y = self.bc3*self.bd1 - self.bc1*self.bd3
        self.n2z = self.bc1*self.bd2 - self.bc2*self.bd1

    def angle(self):
        self.n1()
        self.n2()

        n1n2 = self.n1x*self.n2x + self.n1y*self.n2y + self.n1z*self.n2z

        abs_n1 = math.sqrt(self.n1x**2 + self.n1y**2 + self.n1z**2)
        abs_n2 = math.sqrt(self.n2x**2 + self.n2y**2 + self.n2z**2)

        cos_phi = abs(n1n2) / (abs_n1 * abs_n2)
        ans = math.degrees(math.acos(cos_phi))

        return f"{ans:.2f}"


a1,a2,a3 = map(int , input().split())
b1,b2,b3 = map(int , input().split())
c1,c2,c3 = map(int , input().split())
d1,d2,d3 = map(int , input().split())
A = PHI(a1,a2,a3,b1,b2,b3,c1,c2,c3,d1,d2,d3)
print(A.angle())


```

```python
#W8a13
import math
class Area:
  def __init__(self,a,b,c):
    self.a = a
    self.b = b
    self.c = c
  def check(self):
    if self.a + self.b > self.c and self.b + self.c > self.a and self.a + self.c > self.b :
      return True
    else:
      return False
  def ans(self):
    self.a = a
    self.b = b
    self.c = c
    p = (a + b + c) // 2
    ans = math.sqrt(p*(p-a)*(p-b)*(p-c))
    return f"{ans:.2f}"
a,b,c = map(int,input().split())
Tri = Area(a,b,c)
if Tri.check() == False :
  print("Invalid")
else :
  print(Tri.ans())

```
