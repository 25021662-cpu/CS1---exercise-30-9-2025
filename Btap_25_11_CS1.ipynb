```python
#Max_of_two
def max_of_two(x,y):
    if x>y:
        return x
    return y
n = int(input())
while n > 0 :
  a,b = map(int, input().split())
  print(max_of_two(a,b))
  n -= 1
```

```python
#Swap
def swap(a,b):
  a,b = b,a
  return a,b
n = int(input())
while n > 0 :
  a,b = map(int, input().split())
  print(swap(a,b))
  n -= 1
```

```python
#Isprime
from math import sqrt

def Isprime(n):
    if n < 2:
        return False
    for i in range(2, int(sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

n = int(input())
while n > 0:
    a = int(input())
    print(Isprime(a))
    n -= 1
```

```python
#Perfectnumber
def Perfectnumber(n):
  sum = 0
  for i in range(1,n):
    if n % i == 0:
      sum += i
  if sum == n:
    return True
  return False
n = int(input())
print(Perfectnumber(n))
```

```python
#giathua
def giaithua(n):
  if n == 0:
    return 1
  return n * giaithua
n = int(input())
print(giaithua(n))
```

```python
#Maytinh
def computer(a, b, s):
    if s == '+':
        return f"{(a + b):.2f}"
    if s == '-':
        return f"{(a - b):.2f}"
    if s == '*':
        return f"{(a * b):.2f}"
    if s == '/':
        return f"{(a / b):.2f}"

n = int(input())
while n > 0:
    a, s, b = input().split()
    a = float(a)
    b = float(b)
    print(computer(a, b, s))
    n -= 1
1
```

```python
#hamming
def np(a):
    if a == 0:
        return "0"
    n = ""
    while a > 0:
        n = str(a % 2) + n
        a //= 2
    return n

def hamming(a, b):
    sa = np(a)
    sb = np(b)
    k = max(len(sa), len(sb))
    h = min(len(sa), len(sb))
    cnt = 0
    for i in range(1,h +1 ):
        if sa[-i] != sb[-i]:
            cnt += 1
    return cnt + (k - h)

a, b = map(int, input().split())
print(hamming(a, b))
```
