```python
#W7A1
def binary_search(arr, target):
    left = 0
    right = len(arr) - 1
    ans = -1

    while left <= right:
        midle = (left + right) // 2
        if arr[midle] == target:
            ans = midle
            right = midle - 1
        elif arr[midle] < target:
            left = midle + 1
        else:
            right = midle - 1

    return ans


arr = list(map(int, input().split()))
target = 9
print(binary_search(arr, target))
```

```python

#W7A2
def fist_pos(arr,target):
  left = 0
  right = len(arr) - 1
  ans = -1
  while(left <= right):
    mid = (left + right) // 2
    if arr[mid] == target:
      ans = mid
      right = mid - 1
    elif arr[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return ans
def final_pos (arr,target):
  left = 0
  right = len(arr) - 1
  ans = -1
  while(left <= right):
    mid = (left + right) // 2
    if arr[mid] == target:
      ans = mid
      left = mid + 1
    elif arr[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return ans
def count_occurrences (arr,target):
  l = fist_pos(arr,target)
  if l == -1 :
    return 0
  r = final_pos(arr,target)
  return r - l + 1
arr = list(map(int, input().split()))
target = 2
print(count_occurrences(arr, target))
```

```python
#W7a7
nums = list(map(int, input().split()))
n = len(nums)

dp = [1] * n

for i in range(n):
    for j in range(i):
        if nums[j] < nums[i]:
            dp[i] = max(dp[i], dp[j] + 1)

print(max(dp))
```

```python
#W7a9
def count (a,x):
  cnt = 0
  for i in range(x,len(a)):
    if a[x] > a[i]:
      cnt+=1
  return cnt
a = list(map(int,input().split()))
b = []
for i in range(len(a)):
  k = count(a,i)
  b.append(k)
print(b)
```

```python
#w7a10
def prefix_sum(a):
    b = []
    b.append(a[0])
    for i in range(1, len(a)):
        b.append(b[i-1] + a[i])
    return b

a = list(map(int, input().split()))
key = 2
c = prefix_sum(a)

mod = []

for i in range(len(c)):
    for j in range(i, len(c)):
        if i == 0:
            ans = c[j] % key
        else:
            ans = (c[j] - c[i-1]) % key
        mod.append(ans)

print(max(mod))
```

```python
#w7a11
def compare(s, x):
    i = 0
    while i < len(s):
        if s[i].isalpha() == False:
            s = s[:i] + s[i+1:]
        else:
            i += 1

    if len(s) != len(x):
        return 0
    else:
        cnt = 0
        for i in range(len(s)):
            if s[i] == x[i]:
                cnt += 1
            else:
                return 0
        if cnt == len(s):
            return 1


s = input()
x = 'die'
words = s.split()
cnt = 0

for i in words:
    k = str(i)
    if compare(k, x) == 1:
        cnt += 1

print(cnt)
```

```python
#w7a14
def find7(a):
  pos = []
  for i in range(len(a)):
    if a[i] == 7 :
      pos.append(i)
  if len(pos) == 0 :
    return "Not found"
  return pos
a = list(map(int,input().split()))
print(*find7(a))

```
