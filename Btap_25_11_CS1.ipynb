```python
#Max_of_two
def max_of_two(x,y):
    if x>y:
        return x
    return y
n = int(input())
while n > 0 :
  a,b = map(int, input().split())
  print(max_of_two(a,b))
  n -= 1
```

```python
#Swap
def swap(a,b):
  a,b = b,a
  return a,b
n = int(input())
while n > 0 :
  a,b = map(int, input().split())
  print(swap(a,b))
  n -= 1
```

```python
#Isprime
from math import sqrt

def Isprime(n):
    if n < 2:
        return False
    for i in range(2, int(sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

n = int(input())
while n > 0:
    a = int(input())
    print(Isprime(a))
    n -= 1
```

```python
#Perfectnumber
def Perfectnumber(n):
  sum = 0
  for i in range(1,n):
    if n % i == 0:
      sum += i
  if sum == n:
    return True
  return False
n = int(input())
print(Perfectnumber(n))
```

```python
#
a = list(map(int,input().split()))
def find(n):
  for i in range(0,len(a)-1) :
    if n == a[i]:
      return i
  return -1
n = int(input())
print(find(n))
```

```python
def giaithua(n):
    if n == 0:
        return 1
    return n * giaithua(n - 1)

n = int(input())
print(giaithua(n))
```
