```python
#w6a9
def nhap_dict():
    d = {}
    pairs = input().split()
    for p in pairs:
        k, v = p.split(':')
        d[k] = int(v)
    return d

d1 = nhap_dict()
d2 = nhap_dict()
d3 = {}

for k in d1:
    d3[k] = d1[k]

for k in d2:
    if k in d3:
        d3[k] += d2[k]
    else:
        d3[k] = d2[k]

for k in sorted(d3):
    print(k, d3[k])
```

```python
#w6a11
t = tuple(map(int, input().split()))

odd = []
even = []

for x in t:
    if x % 2 == 0:
        even.append(x)
    else:
        odd.append(x)

t1 = tuple(odd)
t2 = tuple(even)

print(t1)
print(t2)
```

```python
#w6a12
a = list(map(int,input().split()))
cnt = [0]*len(a)
for i in a :
  cnt[i] += 1
k = max(cnt)
for i in range(len(cnt)):
  if cnt[i] == k :
    print(i, end=' ')
    break
```

```python
#w6a15
pairs = input().split()
scores = int(input())
d = {}
ans = {}

for p in pairs:
    k, v = p.split(':')
    d[k] = int(v)

for k, v in d.items():
    if v > scores:
        ans[k] = v

print(ans)

```
